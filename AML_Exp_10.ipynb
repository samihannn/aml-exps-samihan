```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules

df=pd.DataFrame({
'milk':[1,0,1,1],
'bread':[1,1,1,0],
'butter':[0,1,1,1]
})

df=df.astype(bool)

freq=apriori(df,min_support=0.2,use_colnames=True)
rules=association_rules(freq,metric="confidence",min_threshold=0.5)

print(rules)
```
